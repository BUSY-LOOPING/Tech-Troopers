```python
#make the above flow into a function
def test_and_return_best_model(x_col):
    #using poly
    degree = 2
    coefs = np.polyfit(df[x_col], global_y, degree)
    poly = np.poly1d(coefs)
    
    predicted_y_poly = poly(df[x_col])
    residuals_poly = global_y - predicted_y_poly
    
    #using linear regression 
    slope, intercept, r_value, p_value, std_err = linregress(df[x_col], global_y)
    predicted_y_lin = intercept + slope * df[x_col]
    residuals_lin = global_y - predicted_y_lin

    if(residuals_lin.std() < residuals_poly.std()):
        df_res = pd.DataFrame({'predicted': predicted_y_lin, 'residuals': residuals_lin})
        return df_res

    df_res =  pd.DataFrame({'predicted': predicted_y_poly, 'residuals': residuals_poly})
    return df_res
```

```python
predicted_df_results = []
for col_name, corr_value in highly_corr.items():
    df_res = test_and_return_best_model(col_name)
    df_res[target_column] = global_y
    df_res['time'] = x
    predicted_df_results.append(df_res)

    #plot results
    print('predicting using col: {c}, corr: {c_v}'.format(c = col_name, c_v = corr_value))
    plt.plot(x, global_y, label = 'actual')
    plt.plot(x, df_res['predicted'], label='predicted')
    plt.legend()
    plt.show()

    plt.plot(x, df_res['residuals'], label='diff')
    plt.legend()
    plt.show()
```

```python
#get outliers using IQR

def calculate_IQR(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)

    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify outliers
    outliers_iqr = data[(data < lower_bound) | (data > upper_bound)]
    return outliers_iqr
```

```python
idx = 0
outlier_time_dict = {}

for col_name, corr_value in highly_corr.items():
    df_ress = predicted_df_results[idx]
    outliers = calculate_IQR(df_ress['residuals'])

    outliers_time = df_ress.iloc[outliers.index]['time']
    outlier_time_dict[col_name] = outliers_time

    idx += 1

outlier_time_dict
```

```python
#map all outliers
plt.figure(figsize=(20, 15))
for col_name, time_series in outlier_time_dict.items():
    local_y = df.iloc[time_series.index][target_column]
    plt.scatter(time_series, local_y, label=col_name)

plt.plot(x, global_y, label='actual motor data')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
x = df.drop(columns=target_column).values
y = global_y.values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

model = Sequential()
model.add(Dense(64, input_dim=25, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=1)
```

```python
y_pred = model.predict(x_test)
mse = model.evaluate(x_test, y_test, verbose=0)
mse
```
